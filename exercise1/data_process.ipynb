```python
import json
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./data/olympics.csv')
```

```python
data
```

### Process for map visualization

```python
for_map = data.groupby(['Country', 'Country Code']).size().reset_index(name='total')
```

```python
for_map.rename(columns={"Country": "name", "Country Code": "code", "total":"pop"}, inplace=True)
```

```python
for_map.to_csv("for_map.csv", index=False)
```

### Process for stacked bar chart

```python
total_count_df = data.groupby(['Country', 'Sport']).size().reset_index(name='total')
```

```python
total_count_df
```

```python
country_list = ['Canada','United States','Norway','Finland','Sweden','Germany','Switzerland']
```

```python
top_6_mask = total_count_df.Country.isin(country_list)
```

```python
top_6 = total_count_df[top_6_mask]
```

```python
top_6
```

```python
stacked_top_6 = top_6.pivot(index='Sport', columns='Country', values='total').reset_index()
stacked_top_6
```

```python
stacked_top_6 = stacked_top_6.fillna(0)
stacked_top_6.iloc[:,1:] = stacked_top_6.iloc[:,1:].astype(int)
stacked_top_6
```

```python
stacked_top_6.to_csv('stacked_country.csv', index=False)
```
